#### ==========================================================================================================

#### STATISTIK DATA 

```python
# Statistik tentang data
import pandas as pd
```

```python
kapasitas_sekolah = pd.read_csv('kapasitassekolahperikanan.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
print ("             DATA KAPASITAS SEKOLAH PERIKANAN DI SELURUH INDONESIA TAHUN 2003-2012")

kapasitas_sekolah
```

```python
# Indeks pada Tabel Data
print ("INDEKS PADA TABEL DATA")
kapasitas_sekolah.columns
```

```python
# Banyak Sekolah Perikanan
print ("BANYAK SEKOLAH PERIKANAN DI SELURUH INDONESIA ADALAH")
kapasitas_sekolah.SekolahPerikanan.count()
```

```python
# Minat Pendaftar untuk Melanjutkan ke Sekolah Perikanan
print ("MINAT PENDAFTAR UNTUK MELANJUTKAN KE SEKOLAH PERIKANAN")
print ()
s = kapasitas_sekolah.JumlahPendaftar.sum()
mi = kapasitas_sekolah.JumlahPendaftar.min()
ma = kapasitas_sekolah.JumlahPendaftar.max()

print ("Banyak Pendaftar di Sekolah Perikanan sejak Tahun 2003 hingga 2012 adalah", "calon siswa")
print ("Dari Data, Diketahui bahwa Pendaftar Paling Sedikit sebanyak", mi, "calon siswa")
print ("Dari Data, Diketahui bahwa Pendaftar Paling Banyak sejumlah", ma, "calon siswa")
```

```python
# Rata-Rata dan Nilai Tengah
me = kapasitas_sekolah.mean()
md = kapasitas_sekolah.median()

print ("RATA-RATA")
print(me)
print ()
print ("NILAI TENGAH")
print (md)
```

```python
# Rata-Rata Banyak Pendaftar dan Kapasitas Daya Tampung Sekolah Setiap Tahun Ajaran
print ("RATA-RATA BANYAK PENDAFTAR DAN KAPASITAS DAYA TAMPUNG SEKOLAH PERIKANAN SETIAP TAHUN")
kapasitas_sekolah.groupby('TahunAjaran').mean()
```

#### ==========================================================================================================

#### VISUALISASI DATA 

```python
ax = kapasitas_sekolah.groupby('TahunAjaran').mean()['JumlahPendaftar']
ax = ax.plot(kind='bar', title ="Rata-Rata Pendaftar Setiap Tahun", legend=True, fontsize=10)
ax.set_xlabel("Tahun Ajaran", fontsize=12)
ax.set_ylabel("Jumlah Pendaftar", fontsize=12)
```

```python
ax = kapasitas_sekolah.groupby('TahunAjaran').mean()['JumlahKapasitas']
ax = ax.plot(kind='bar', title ="Rata-Rata Kapasitas Sekolah Setiap Tahun", legend=True, fontsize=10)
ax.set_xlabel("Tahun Ajaran", fontsize=12)
ax.set_ylabel("Kapasitas Sekolah", fontsize=12)
```
